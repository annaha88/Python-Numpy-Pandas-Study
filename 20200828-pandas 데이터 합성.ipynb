```python
import numpy as np
import pandas as pd
from pandas import DataFrame, Series
# 위와 같이 DataFrame, Series import 해두면, pd. 굳이 써주지 않아도 된다. 
```

```python
data = {
        'key':['b','b','a','c','a','a','b']
      , 'data1':range(7)
}
df1 = DataFrame(data)
df1
```

```python
data11= {'key': ['a', 'b', 'd'],
                 'data2': range(3)}
df2 = DataFrame(data11)
df2
```

```python
pd.merge(df1,df2)
```

```python
pd.merge(df1, df2, on='key') #inner join
```

```python
df3=DataFrame({'lkey': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                 'data1': range(7)})
```

```python
df4=DataFrame({'rkey': ['a', 'b', 'd'],
                 'data2': range(3)})
```

```python
pd.merge(
df3,df4, left_on='lkey'
    ,right_on='rkey'
)
```

```python
#컬럼명이 다를 경우 lkey,rkey 따로 잡아주기.
```

```python
left = DataFrame({'key1': ['foo', 'foo', 'bar'],
                  'key2': ['one', 'two', 'one'],
                  'lval': [1, 2, 3]})
```

```python
right = DataFrame({'key1': ['foo', 'foo', 'bar', 'bar'],
                   'key2': ['one', 'one', 'one', 'two'],
                   'rval': [4, 5, 6, 7]})
```

```python
pd.merge(left, right, on =['key1','key2'])
```

```python
pd.merge(left, right, on =['key1','key2'])
```

```python
pd.merge(left, right, on =['key1','key2'], how='outer')
```

```python
pd.merge(left, right, on =['key1'], how='outer')
```

```python
pd.merge(left, right, on ='key1', suffixes=('_left','_right'), how='outer')
```

```python
left1 = DataFrame({'key': ['a', 'b', 'c'],
                   'value': range(3)})
left1
```

```python
right1 = DataFrame({'group_val': [3.5, 7]}, index=['a', 'b'])
right1
```

```python
pd.merge(left1, right1, left_on='key', right_index=True, how='outer')
```

```python
pd.merge(left1, right1, left_on='key', right_index=True, how='inner')
```

```python
pd.merge(left1, right1, left_on='key', right_index=True, how='left')
```

```python
indexes = [['Nevada','Nevada','Ohio','Nevada','Ohio','Ohio'],['2001','2000','2000','2001','2001','2001']]
indexes
```

```python
lefth=DataFrame(
    {'key': ['Nevada','Ohio','Ohio']
     , 'year': ['2001','2001','2002']
     , 'data' : [2,3,4]}
)

lefth
```

```python
righth=DataFrame(
data = np.arange(0,12).reshape(6,2)
,index = indexes
, columns=['data1','data2']

)

righth
```

```python
pd.merge(lefth,righth, left_on=['key','year'], right_index = True, how='outer')
```

```python
left1 = DataFrame(
    data=[[1,2],[3,4]]
    , index = ['a','b']
    , columns= ['Mi','Al']
)
left1
```

```python
right1 = DataFrame(
    data=[[5,6],[7,8]]
    , index = ['b','c']
    , columns= ['Ne','Nw']
)
right1
```

```python
pd.merge(left1, right1, left_index=True, right_index=True, how= 'outer')
```

```python
right1.join(left1, how='outer')
```

```python
pd.merge(left1, right1, left_index=True, right_index=True)
```

```python
right1.join(left1)       
#이방법 쓸때는 무조건 outer 조인밖에 안된다.
# join 방법은 디폴트로 인덱스 가져온다. 
```

```python
right1.join(left1, on = '컬럼명')       
#이방법 쓸때는 무조건 outer 조인밖에 안된다.
# join 방법은 디폴트로 인덱스 가져온다. 
```
