모듈 이용한 방법

```python
# URL 분리 => 합치는 모듈 필요

from urllib.parse import urljoin
```

```python
# 메인페이지 가져올 정보 저장
# 1. 랭킹 : rank[]
# 2. 메인메뉴 : main_menu[]
# 3. 카페이름 : cafa_name[]
# 4. URL 저장 : url_add[]
rank = []
main_menu=[]
cafe_name=[]
url_add=[]
```

```python
# 50개의 맛집 정보
list_soup = soup.find_all('div', class_='sammy')

for item in list_soup:
    # 순위
    rank.append(item.find(class_='sammyRank').get_text())
    # 메뉴, 카페
    tmp_string = item.find(class_='sammyListing').get_text()
    # 정규식이 적용된 리스트
    re_string = re.split(('\n|\r\n'), tmp_string)
    # 메뉴추가
    main_menu.append(re_string[0])
    # 카페 이름 추가
    cafe_name.append(re_string[1])
    #url 추가 (urljoin 쓰면 겹치는 문제가 없어서 좋다.)
    tmp_url = urljoin(url_base, item.find('a')['href'])
    url_add.append(tmp_url)
```

```python
# 위 데이터로 데이터프레임 만들기(딕셔너리 구조 이용)
data= {
      'Rank': rank
    , 'Menu': main_menu
    , 'Cafe': cafe_name
    , 'URL' : url_add
}
df = pd.DataFrame(data)
df.head()
```

```python
df = pd.DataFrame(data, columns=['Rank','Cafe','Menu','URL']) # 컬럼 순서 바꾸고 싶으면, 컬럼즈만 주기.
df.head()
```

```python
df.to_csv(
    './03. best_sandwiches_list_chicago.csv'
    ,sep=','
    ,encoding='utf-8'
)
```
